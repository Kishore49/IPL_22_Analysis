Number of matches won by a team

```python
import pandas as pd
import plotly.express as px

# Read the CSV file into a DataFrame
df = pd.read_csv('ipl_2022.csv')

# Calculate the number of matches won by each team
matches_won = df['match_winner'].value_counts().reset_index()
matches_won.columns = ['Team', 'Matches Won']

# Create a bar chart for number of matches won by each team
figure = px.bar(matches_won, x='Team', y='Matches Won',
                title="Number of Matches Won in IPL 2022",
                labels={'Team': 'Team', 'Matches Won': 'Number of Matches Won'},
                color='Team',  # Color bars based on team
                text='Matches Won'  # Display the exact count on each bar
                )
figure.update_traces(texttemplate='%{text}', textposition='outside')  # Set the position of the text labels

# Update layout
figure.update_layout(
    xaxis_title='Team',  # Set the x-axis title
    yaxis_title='Number of Matches Won',  # Set the y-axis title
    showlegend=False  # Hide the legend
)

figure.show()
```

Number of matches won by defending and chasing

```python
import pandas as pd
import plotly.graph_objects as go

# Read the CSV file into a DataFrame
df = pd.read_csv('ipl_2022.csv')

# Map won_by values
df["won_by"] = df["won_by"].map({"Wickets": "Chasing", "Runs": "Defending"})

# Count values
won_by = df["won_by"].value_counts()

label = won_by.index
counts = won_by.values
colors = ['gold', 'silver']

# Create plot
fig = go.Figure(data=[go.Pie(labels=label, values=counts, hole=0.3, pull=[0.1, 0])])
fig.update_layout(title='Number of Matches Won By Defending Or Chasing',
                  annotations=[dict(text=f'{label[i]}: {counts[i]} ({counts[i]/sum(counts)*100:.2f}%)',
                                    x=0.5, y=0.5-0.05*i, font_size=18, showarrow=False) for i in range(len(label))],
                  showlegend=True,
                  legend=dict(title='Winning Method', orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1))

fig.update_traces(marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

Top scorers in the IPL

```python
import pandas as pd
import plotly.express as px

# Read the CSV file into a DataFrame
df = pd.read_csv('ipl_2022.csv')

# Check if 'top_scorer' column exists in the DataFrame
if "top_scorer" in df.columns:
    # Get the count of each player
    player_counts = df['top_scorer'].value_counts().reset_index()
    player_counts.columns = ['top_scorer', 'count']

    # Create the bar chart
    fig = px.bar(player_counts, x='top_scorer', y='count',
                 title="Top Scorers in IPL 2022",
                 labels={'count': 'Total Count'},
                 color='top_scorer')

    # Enable clicking on bars to display player details
    fig.update_traces(hovertemplate='Player: %{x}<br>Total Count: %{y}')

    fig.show()
else:
    print("'top_scorer' column not found in the DataFrame.")
```

```python
import pandas as pd
import plotly.express as px

# Read the CSV file into a DataFrame
df = pd.read_csv('ipl_2022.csv')

# Check if 'player_of_the_match' column exists in the DataFrame
if "player_of_the_match" in df.columns:
    # Get the count of player_of_the_match awards for each player
    player_of_the_match_counts = df['player_of_the_match'].value_counts().reset_index()
    player_of_the_match_counts.columns = ['player_of_the_match', 'count']

    # Create the bar chart
    fig = px.bar(player_of_the_match_counts, x='player_of_the_match', y='count',
                 title="Most Player of the Match Awards",
                 labels={'player_of_the_match': 'Player of the Match', 'count': 'Number of Awards'},
                 color='player_of_the_match',
                 hover_data={'player_of_the_match': False, 'count': True})

    # Add player name to the legend
    fig.update_traces(showlegend=True)

    fig.show()
else:
    print("'player_of_the_match' column not found in the DataFrame.")
```

```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

# Read the CSV file into a DataFrame
data = pd.read_csv('ipl_2022.csv')

# Get unique stadium names
stadiums = data['venue'].unique()

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Select a Stadium"),
    dcc.Dropdown(
        id='stadium-dropdown',
        options=[{'label': stadium, 'value': stadium} for stadium in stadiums],
        value=None,
        placeholder="Select a stadium"
    ),
    html.Div(id='output-container')
])

# Define the callback to update the output based on the selected stadium
@app.callback(
    Output('output-container', 'children'),
    [Input('stadium-dropdown', 'value')]
)
def update_output(selected_stadium):
    if selected_stadium is not None:
        # Filter the DataFrame based on the selected stadium
        filtered_data = data[data['venue'] == selected_stadium]

        # Create a list of matches at the selected stadium
        matches = filtered_data['team1'] + ' vs ' + filtered_data['team2']

        # Create a formatted output
        output = html.Div([
            html.H3(f"Matches at {selected_stadium}:"),
            html.Ul([html.Li(match) for match in matches])
        ])
        return output

    # If no stadium is selected, display a message
    return html.Div("Select a stadium to see matches")

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import pandas as pd
import plotly.graph_objects as go

# Read the CSV file into a DataFrame
data = pd.read_csv('ipl_2022.csv')

# Calculate total matches played in each stadium
total_matches = data.groupby('venue')['match_winner'].count()

# Create a figure
figure = go.Figure()

# Add bars for each team's matches and wickets
for team in data['team1'].unique():
    team_data = data[(data['team1'] == team) | (data['team2'] == team)]
    venue_wickets = team_data.groupby('venue')['first_ings_wkts'].sum() + team_data.groupby('venue')['second_ings_wkts'].sum()
    figure.add_trace(go.Bar(
        x=venue_wickets.index,
        y=venue_wickets.values,
        name=f'{team} Matches Played',
        text=venue_wickets.values,
        hoverinfo='x+y+text',  # Display venue, matches played, and wickets on hover
    ))

# Add total matches as annotations at the top of each bar
for venue, total_match in total_matches.items():
    figure.add_annotation(
        x=venue,
        y=total_match + 1,  # Place annotation slightly above the top of the bar
        text=f'Total Matches: {total_match}',
        showarrow=False,
        font=dict(size=10)
    )

# Update layout
figure.update_layout(
    barmode='stack',  # Stack bars on top of each other
    xaxis_tickangle=-45,  # Rotate x-axis labels for better readability
    title='Matches Played and Wickets Taken by Team in Each Venue',
    xaxis_title='Venue',
    yaxis_title='Number of Matches/Wickets',
    legend_title='Teams',
    hovermode='x unified'  # Display hover information for all bars when hovering over a bar
)

figure.show()
```
